# **Assignment 1 – Simple Linear Regression**
**Goal:** Understand the basics of regression by creating your own dataset.

### Steps:
1. Formulate your own equation in the form:  
   $
   y = m x + c
   $
   
   Example: $y = 2x + 3$
2. Generate X values (at least 100 numbers between your chosen range).
3. Generate Y values using your equation.
4. Add small random noise to Y to make it realistic.
5. Split data into train and test splits
6. Fit a Simple Linear Regression model.
7. Print learned coefficient and intercept.
8. Split the data into train and test sets (80%-20%).
9. Evaluate both train and test using MAE, MSE, R² score.
> Write your notes about model performance and whether the model suffers from overfitting or underfitting.
10. Visualize the points and the fitted line.


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
m=2
c=3
x=np.random.randint(0,50,(100,1))
x
```

```python
# y=mx+c: y=2x+3
y= (m*x)+c
y[0:5]
```

```python
#add noise to y:
noise=np.random.randint(-5, 5, y.shape)
y_noise=y+noise
```

```python
#plot x,y:
plt.scatter(x,y_noise)
plt.title('correlation between x and y')
plt.xlabel("X")
plt.ylabel("y")
```

```python
#split:
X_train,X_test,y_train,y_test=train_test_split(x,y_noise,test_size=0.2,random_state=42)
```

```python
#fit
model=  LinearRegression()
model.fit(X_train,y_train)
```

```python
#Print learned coefficient and intercept:
coeff=model.coef_
intercept=model.intercept_
print(f'learned coefficient: {coeff}\nlearnedinercept: {intercept} ')
```

-the learned coefficient and the learned intercept are extremely close to the original ones


```python
#8.Split the data into train and test sets (80%-20%): already done in the previous steps
```

```python
#predict:
y_train_pred=model.predict(X_train)
y_test_pred=model.predict(X_test)
```

```python
#evaluate:
# MAE, MSE, R²
mae_train=mean_absolute_error(y_train,y_train_pred)
mae_test=mean_absolute_error(y_test,y_test_pred)

mse_train=mean_squared_error(y_train,y_train_pred)
mse_test=mean_squared_error(y_test,y_test_pred)

r2_train=r2_score(y_train,y_train_pred)
r2_test=r2_score(y_test,y_test_pred)

print(f"Train:\nMAE={mae_train} \nMSE={mse_train} \nR2 score={r2_train}")
print(" ")
print(f"Test:\nMAE={mae_test} \nMSE={mse_test} \nR2 score={r2_test}")
```

insights:

The R² scores for train and test are extremely close.

The errors (MAE & MSE) are small and very close for both sets.

This means the model generalizes well and does not suffer from overfitting or underfitting



```python
plt.scatter(X_train, y_train, color='blue', label='Train Data')
plt.scatter(X_test, y_test, color='red', label='Test Data')

x_line = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)
y_line = model.predict(x_line)
plt.plot(x_line, y_line, color='black', label='Fitted Line')

plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Simple Linear Regression Fit")
```

- no overfitting or underfitting
- The spread of points around the line is minimal, indicating low error.
- the line fits the data very well

# **Assignment 3 – Polynomial Regression**
**Goal:** Model non-linear relationships.

### Steps:
1. Formulate your own equation:
   $
   y = ax^2 + bx + c
   $
2. Generate X values and Y values with noise.
3. Use `PolynomialFeatures` to transform X.
4. Split the data into train and test sets (80%-20%).
5. Fit a Linear Regression model on transformed features.
6. Evaluate both train and test using MAE, MSE, R² score.
> Write your notes about model performance and whether the model suffers from overfitting or underfitting.



```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
a = 2
b = 3
c = 5
x = np.random.randint(-20, 20, (100, 1))
x[0:5]
```

```python
y=(a*x**2)+(b*x)+c
y[0:5]
```

```python
noise=np.random.randn(100, 1)*50
y=y+noise
y[0:5]
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(x)
```

```python
#split
X_train,X_test,y_train,y_test= train_test_split(X_poly,y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#predict:
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
#evaluate:
mae_train=mean_absolute_error(y_train,y_train_pred)
mae_test=mean_absolute_error(y_test,y_test_pred)

mse_train=mean_squared_error(y_train,y_train_pred)
mse_test=mean_squared_error(y_test,y_test_pred)

r2_train=r2_score(y_train,y_train_pred)
r2_test=r2_score(y_test,y_test_pred)

print(f"Train:\nMAE={mae_train} \nMSE={mse_train} \nR2 score={r2_train}")
print(" ")
print(f"Test:\nMAE={mae_test} \nMSE={mse_test} \nR2 score={r2_test}")
```

-  very high R2 scores for both training and testing means it captures the non-linear relationship very well.

- MAE and MSE are very close between train and test indicates the model generalizes well and not overfitting.

- performance is strong

```python
plt.scatter(x, y, color='blue', label='Data Points')
X_sorted = np.sort(x, axis=0)
X_poly_sorted = poly.transform(X_sorted)
y_pred_line = model.predict(X_poly_sorted)

plt.plot(X_sorted, y_pred_line, color='red', linewidth=2, label='Polynomial Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.legend()
```

- polynomial line fits the data

```python
onehot_features=['airline', 'source_city', 'destination_city']
ordinal_features_time=['departure_time', 'arrival_time']

onehot = OneHotEncoder(sparse_output=False, drop=None, handle_unknown="ignore")
ohe_df = pd.DataFrame(onehot.fit_transform(df[onehot_features]),
                      columns=onehot.get_feature_names_out(onehot_features),
                      index=df.index)

#replace original categorical columns with one-hot columns
df = pd.concat([df.drop(columns=onehot_features), ohe_df], axis=1)


for col in ordinal_features_time:
    df[col] = df[col].map({'Early_Morning':1, 'Morning':2, 'Afternoon':3,
    'Evening':4, 'Night':5, 'Late_Night':6})

df.head()
```

```python
df['stops'] = df['stops'].map({'zero': 0,
                               'one': 1,
                               'two_or_more': 2
})
df['stops'].unique()
```

```python
df['class']=df['class'].map({'Economy':0,'Business':1})
```

```python
#frequency encoding for flight
freq_map = df['flight'].value_counts(normalize=True)
df['flight'] = df['flight'].map(freq_map)
```

```python
#split data
x=df.drop(columns=['price'])
y=df['price']
```

```python
X_train, X_test, y_train, y_test= train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
#scaling:
cols_to_scale = ['duration', 'days_left', 'departure_time', 'arrival_time', 'stops']

#log transform for skewed duration
X_train['duration'] = np.log1p(X_train['duration'])
X_test['duration'] = np.log1p(X_test['duration'])

#scale continuous features
scaler = RobustScaler()
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()
X_train_scaled[cols_to_scale] = scaler.fit_transform(X_train[cols_to_scale])
X_test_scaled[cols_to_scale] = scaler.transform(X_test[cols_to_scale])
```

```python
pipeline = Pipeline([
    ("scaler", RobustScaler()),
    ("model", LinearRegression())
])

#Parameter grid
param_grid = [
    {"model": [LinearRegression()]},
    {"model": [Ridge()], "model__alpha": [0.01, 0.1, 1, 10, 100]},
    {"model": [Lasso(max_iter=5000)], "model__alpha": [0.01, 0.1, 1, 10, 100]}
]

#Grid Search
grid = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    scoring="r2",
    cv=5,
)
grid.fit(X_train_scaled, y_train)
```

```python
#best model
best_model = grid.best_estimator_
print(f"Best Model: {best_model}")
print(f"Best CV R²: {grid.best_score_}")

#evaluate on train/test
y_train_pred = best_model.predict(X_train_scaled)
y_test_pred = best_model.predict(X_test_scaled)

#train
print("\nPerformance:")
print(f"Train -> MAE: {mean_absolute_error(y_train, y_train_pred):}, "
      f"MSE: {mean_squared_error(y_train, y_train_pred):}, "
      f"RMSE: {np.sqrt(mean_squared_error(y_train, y_train_pred)):}, "
      f"R²: {r2_score(y_train, y_train_pred):}")

#test
print(f"Test  -> MAE: {mean_absolute_error(y_test, y_test_pred):}, "
      f"MSE: {mean_squared_error(y_test, y_test_pred):}, "
      f"RMSE: {np.sqrt(mean_squared_error(y_test, y_test_pred))}, "
      f"R²: {r2_score(y_test, y_test_pred):}")
```

- RobustScaler handles outliers well as the data has both skewness & outliers

- Lasso Regression made the best performance

- Best CV R²: 0.911 : Very good fit during cross-validation

- the scores for training and test indicates no overfit or underfit as they are very close

- If the train score was much higher than the test score, it would suggest overfitting.  
- If both scores were low, it would suggest*underfittin*
